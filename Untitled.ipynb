```python
import xgboost as xgb
```

```python
#通常
train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size=0.3, shuffle=True) 

#球種割合揃え
#train_x, test_x, train_y, test_y = train_test_split(X2, Y2, test_size=0.3, shuffle=True)

```

```python
dtrain = xgb.DMatrix(train_x, label=train_y)
dtest = xgb.DMatrix(test_x, label=test_y)  
```

```python
#パラメータの設定
param = {'max_depth': 12, 'eta': 0.3, 'gamma': 10,'colsample':0.5, 'objective':'multi:softprob',
         'num_class':8, 'eval_metric':'mlogloss','tree_method':'gpu_hist'}  
'''
パラメータ名　意味
max_depth　木の最大深度
eta　学習率
objective　学習目的
num_class　クラス数

'''
evallist = [(dtest, 'eval'), (dtrain, 'train')] 
```

```python
num_round = 1000  
bst = xgb.train(param, dtrain, num_round, evallist, early_stopping_rounds=10)
```

```python
#モデルの適用
pred = bst.predict(dtest)
#Pandasに変更
pred_p = pd.DataFrame(pred) 
#球種のone-hot化
test_y_onehot = pd.get_dummies(test_y)
```

```python
from sklearn.metrics import accuracy_score  
 
score = accuracy_score(test_y, pred)  
print('score:{0:.4f}'.format(score))
```
